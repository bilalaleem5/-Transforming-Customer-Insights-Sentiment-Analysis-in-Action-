```python
import json
import re
import statistics


# Specify the path to the JSON file
data_path = "Cell_Phones_and_Accessories_5.json"



# Function to load JSON data from a file
def json_data(data_path):

    # Initialize an empty list to store parsed data

    data = []



    # Open the JSON file in read mode
    with open(data_path, 'r', encoding='utf-8') as file:

        # Iterate over each line in the file
        for line in file:

            # Load JSON data from the current line and append it to the list
            data.append(json.loads(line))

    # Return the list containing parsed JSON data
    return data

# Call the json_data function to obtain the parsed data
parsed_data = json_data(data_path)


# Define a set of punctuation characters
punctuation = set("!\"#$%&'()*+,-./:;<=>?@[\\]^_`{|}~")

# Create a translation table to remove punctuation characters
translator = str.maketrans("", "", "".join(punctuation))

# Function to load JSON data from a file
def json_data(data_path):

    data = []

    with open(data_path, 'r', encoding='utf-8') as file:

        for line in file:
 
            data.append(json.loads(line))
 
    return data

# Function to preprocess reviewText data
def preprocess_data(data):

    # Define a set of stopwords to be removed from the text
    stopwords_list = set([
        "i", "me", "my", "myself", "we", "our", "ours", "ourselves", "you", "your", "yours", "yourself", "yourselves", "he", "him", "his", "himself", "she", "her", "hers", 
        "herself", "it", "its", "itself", "they", "them", "their", "theirs", "themselves", "what", "which", "who", "whom", "this", "that", "these", "those", "am", "is", "are", 
        "was", "were", "be", "been", "being", "have", "has", "had", "having", "do", "does", "did", "doing", "a", "an", "the", "and", "but", "if", "or", "because", "as", "until", 
        "while", "of", "at", "by", "for", "with", "about", "against", "between", "into", "through", "during", "before", "after", "above", "below", "to", "from", "up", "down", 
        "in", "out", "on", "off", "over", "under", "again", "further", "then", "once", "here", "there", "when", "where", "why", "how", "all", "any", "both", "each", "few", "more", 
        "most", "other", "some", "such", "no", "nor", "not", "only", "own", "same", "so", "than", "too", "very", "s", "t", "can", "will", "just", "don", "should", "now"])

    processed_data = []

    # Iterate through each dictionary in the input data
    for dictionary in data:

        # Check if 'reviewText' key is present in the dictionary
        if 'reviewText' in dictionary:

            # Get the text to process
            text_to_process = dictionary['reviewText']

            # Split the text into words
            words = text_to_process.split()

            # Filter out stopwords and punctuation, convert to lowercase
            filtered_words = [word.lower() for word in words if word.lower() not in stopwords_list and word.lower() not in punctuation]

            # Join the filtered words to form the processed text
            processed_text = ' '.join(filtered_words)

            # Append the processed text to the result list
            processed_data.append(processed_text)

    # Return the list of processed texts
    return processed_data

# Load JSON data from the specified file
json_data_fun = json_data(data_path)

# Preprocess the reviewText data
preprocess_data_result = preprocess_data(json_data_fun)

# Display the result of the preprocessing
preprocess_data_result


from collections import Counter

# Load JSON data from the specified file
reviews_data = json_data(data_path)

# Function to remove punctuation and stopwords from a given text
def remove_punctuation_and_stopwords(text):

    # Define a set of stopwords to be removed from the text
    stopwords = set([
        "i", "me", "my", "myself", "we", "our", "ours", "ourselves", "you", "your", "yours", "yourself",
        "yourselves", "he", "him", "his", "himself", "she", "her", "hers", "herself", "it", "its", "itself",
        "they", "them", "their", "theirs", "themselves", "what", "which", "who", "whom", "this", "that", "these",
        "those", "am", "is", "are", "was", "were", "be", "been", "being", "have", "has", "had", "having", "do",
        "does", "did", "doing", "a", "an", "the", "and", "but", "if", "or", "because", "as", "until", "while",
        "of", "at", "by", "for", "with", "about", "against", "between", "into", "through", "during", "before",
        "after", "above", "below", "to", "from", "up", "down", "in", "out", "on", "off", "over", "under", "again",
        "further", "then", "once", "here", "there", "when", "where", "why", "how", "all", "any", "both", "each",
        "few", "more", "most", "other", "some", "such", "no", "nor", "not", "only", "own", "same", "so", "than",
        "too", "very", "s", "t", "can", "will", "just", "don", "should", "now"
    ])

    # Define a set of punctuation characters
    punctuation = set("!\"#$%&'()*+,-./:;<=>?@[\\]^_`{|}~")

    # Create a translation table to remove punctuation characters
    translator = str.maketrans("", "", "".join(punctuation))

    # Convert the text to lowercase, remove punctuation, and split into words
    words = text.lower().translate(translator).split()

    # Filter out stopwords
    filtered_words = [word for word in words if word not in stopwords]

    return filtered_words

# Function to calculate the Cumulative Distribution Function (CDF) of a given dataset
def calculate_cdf(data):

    counts = {}

    total_words = 0

    cumulative_frequency = []

    # Iterate through each word in the dataset
    for word in data:

        # Update the count for the current word
        counts[word] = counts.get(word, 0) + 1

        # Increment the total number of words
        total_words += 1

        # Append the cumulative frequency to the list
        cumulative_frequency.append(total_words)

    # Get unique values and calculate CDF
    values = list(counts.keys())

    cdf = [count / total_words for count in cumulative_frequency]

    # Return the unique values and their corresponding CDF
    return values, cdf

# Extract positive reviews and their preprocessed words
positive_reviews = [remove_punctuation_and_stopwords(review['reviewText']) for review in reviews_data if review['overall'] >= 4]

# Extract negative reviews and their preprocessed words
negative_reviews = [remove_punctuation_and_stopwords(review['reviewText']) for review in reviews_data if review['overall'] <= 2]

# Flatten the lists of words in positive and negative reviews

all_positive_words = [word for review in positive_reviews for word in review]

all_negative_words = [word for review in negative_reviews for word in review]

# Calculate CDF for positive and negative reviews

positive_values, positive_cdf = calculate_cdf(all_positive_words)

negative_values, negative_cdf = calculate_cdf(all_negative_words)

# Display the CDF for the first 1000 values in positive and negative reviews

print("Positive Reviews:")

print("Values:", positive_values[:1000])

print("CDF:", positive_cdf[:1000])

print("\nNegative Reviews:")

print("Values:", negative_values[:1000])

print("CDF:", negative_cdf[:1000])


#keywords
from collections import Counter

import re



def extract_keywords(data, threshold=0.01, top_n=5):

    # Flatten the list of words in all reviews
    all_words = [word for review in data for word in review]

    # Count the occurrences of each word
    word_counts = Counter(all_words)
    
    # Calculate the total number of reviews
    total_reviews = len(data)
    
    # Select keywords that appear in at least 'threshold * total_reviews' reviews
    keywords = [word for word, count in word_counts.items() if count >= threshold * total_reviews]
    
    # Sort keywords based on frequency
    keywords.sort(key=lambda x: word_counts[x], reverse=True)

    # Return the top N keywords
    
    return keywords[:top_n]

# Extract keywords from positive and negative reviews

positive_keywords = extract_keywords(positive_reviews)

negative_keywords = extract_keywords(negative_reviews)


# Function for rule-based sentiment analysis
def rule_based_sentiment_analysis(text, threshold=0.0):

    # Convert text to lowercase for case-insensitive matching
    text_lower = text.lower()

    # Count occurrences of positive and negative keywords in the text

    positive_count = sum([len(re.findall(keyword, text_lower)) for keyword in positive_keywords])

    negative_count = sum([len(re.findall(keyword, text_lower)) for keyword in negative_keywords])
    
    # Calculate total count of positive and negative keywords

    total_count = positive_count + negative_count

    # Calculate percentages

    positive_percentage = positive_count / total_count if total_count > 0 else 0

    negative_percentage = negative_count / total_count if total_count > 0 else 0
    
    # Calculate sentiment score

    sentiment_score = positive_percentage - negative_percentage
    
    # Apply threshold to categorize sentiment

    if sentiment_score >= 0.1:

        sentiment_category = "Positive"

    elif sentiment_score <= -0.1:

        sentiment_category = "Negative"

    else:

        sentiment_category = "Neutral"

    return sentiment_score, sentiment_category




# Specify the output file path
output_file_path = "output.txt"


# Open the output file for writing
with open(output_file_path, 'w', encoding='utf-8') as output_file:

    # Iterate through each review in the parsed data
    for review in parsed_data:

        # Preprocess the review text
        text = remove_punctuation_and_stopwords(review['reviewText'])

        # Join the preprocessed text into a single string
        text = ' '.join(text)

        # Perform rule-based sentiment analysis on the text
        sentiment_score, sentiment_category = rule_based_sentiment_analysis(text)
        
        # Write review information and sentiment analysis results to the output file

        output_file.write(f"Review: {text}\n")

        output_file.write(f"Sentiment Score: {sentiment_score}\n")

        output_file.write(f"Sentiment Category: {sentiment_category}\n\n")

 
# Print a message indicating the completion of the process

print(f"Reviews and sentiments have been written to {output_file_path}.")

print('yahuuuuuuuuuuuuuuuu: DONE')

print('THANKYOU :-))))))')



```
